```python
class Nettoyage(pd.core.frame.DataFrame):
    
    def __init__(self, df):
        pd.core.frame.DataFrame.__init__(self,df)
        
    def erase(self,col):
        del self[col]
        return self
    
    def fil_nan(self):
        fillna = self.fillna(method="bfill", inplace=True)
        return fillna
    
    def mapping(self, col, dico):
        self[col] = self[col].map(dico)
        
    def converth(self,col):
        self[col] = round(self[col]*0.3048,2)
        return self[col]
        
    def convertw(self,col):
        self[col] = round(self[col]*0.4535,2)
        return self[col]
    
    def replacestr(self,col, val1, val2):
        self[col] = self[col].str.replace(val1,val2)
        return self[col]
    
    def dico(self, col):
        a = self[col].unique()
        b = range(len(self[col].unique()))
        c = dict(zip(a, b))
        return c
    
    def date(self, col):
        self[col] = pd.to_datetime(self[col])
        
    def astype(self,col,typ):
        try:
            self[col] = self[col].astype(typ)
        except:
            self[col] = self[col].str.replace("'",".").astype(typ)
        return self[col]
        
    def extract(self,col, string):
        self[col] = self[col].str.extract(string)
        return self[col]
    
   
```
